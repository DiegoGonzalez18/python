```python
import random

def run():
    number=False
    random2=random.randint(0,20)
    while not number: #true
        a=int(input('Digita un numero'))
        if(a==random2):
            print('Adivinaste el numero')
            number = True
           
        elif a<random2:
            print('El numero que digitaste es menor al que intestas adivinar')
        else:
            print('El numero que digitaste es mayor al que intestas adivinar')

if __name__=='__main__':
    run()







```

```python

def pali(palabra):
    aux=palabra[::-1]
    if palabra==aux:
        print('Es palindromo')
    else:
        print('No es palindromo')
   
if __name__=='__main__':
   palabra=str(input('Introduce una palabra'))
   pali(palabra)

```
