```python
class Lamp:
    
    _LAMPS = ['''
          .
     .    |    ,
      \   '   /
       ` ,-. '
    --- (   ) ---
         \ /
        _|=|_
       |_____|
    ''',
    '''
         ,-.
        (   )
         \ /
        _|=|_
       |_____|
    ''']
    def __init__(self):
        self._is_turned_on=False
        self._display_image()
    def turn_on(self):
        self._is_turned_on=True
        self._display_image()
    
    def turn_of(self):
        self._is_turned_on=False
    def _display_image(self):
        if  self._is_turned_on:
            print(_LAMPS[0])
        else:
            print(_LAMPS[1])
```
