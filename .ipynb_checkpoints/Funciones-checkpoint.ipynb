```python
import turtle
def main():
    window = turtle.Screen()
    diego= turtle.Turtle()
    make_square(diego)
    turtle.mainloop()
    
    
def make_square(diego):
    length=int(input("Ingrese el tamño del cuadrado: "))
    for i in range(4):
        make_line_and_turn(diego,length)
    
def make_line_and_turn(diego,length):
    diego.forward(length)
    diego.left(90)

    
if __name__ =='__main__':
    main()
```
