```python
import turtle

tortuga = turtle.Turtle()

tortuga.forward(50)
tortuga.right(90)     # Rotate clockwise by 90 degrees

tortuga.forward(50)
tortuga.right(90)

tortuga.forward(50)
tortuga.right(90)

tortuga.forward(50)
tortuga.right(90)

turtle.done()
```
